```
df = pd.read_csv('sports.csv', index_col=0)
```

```
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,10))
v = venn2_wordcloud([set(df['cricket']), set(df['football'])],
                    ax=ax, set_labels=['Cricket', 'Football'],set_edgecolors=['b', 'r'])

# add color
v.get_patch_by_id('10').set_color("red")
v.get_patch_by_id("10").set_alpha(0.4)
v.get_patch_by_id("01").set_color("blue")
v.get_patch_by_id("01").set_alpha(0.4)
v.get_patch_by_id("11").set_color("purple")
v.get_patch_by_id("11").set_alpha(0.4)

ax.set_title("Top Common Words in Cricket and Footbal", fontsize=20)
```
